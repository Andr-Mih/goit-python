```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
A, B, C = [(0,0), (3,3), (6,0)]
```

```python
p = (4, 1)
```

```python
list_x = []
list_y = []
for i in range(0, 10000, 1):
    v = random.choice([A, B, C])
    p_x = (p[0]+v[0])/2
    p_y = (p[1]+v[1])/2
    p = (p_x, p_y)
    list_x.append(p_x)
    list_y.append(p_y)
```

```python
plt.figure(figsize = (10, 5))
```

```python
plt.scatter(list_x, list_y)
plt.show()
```
