```python
python_data.groupby('Должность').mean()
```

```python
new_set = python_data.groupby('Должность')
salary_set = new_set.agg({'Зарплата.в.месяц': ['max', 'min', 'mean']})
salary_set
```

```python
salary_set['Зарплата.в.месяц']['min']
```

```python
def fill_avg_salary(data):
  return (data['Зарплата.в.месяц']['min'] + data['Зарплата.в.месяц']['max'])/2

```

```python
salary_set['avg'] = salary_set.apply(lambda x: fill_avg_salary(x), axis = 1)
salary_set
```
