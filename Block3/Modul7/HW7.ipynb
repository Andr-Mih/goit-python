```python
from sklearn.datasets import  load_wine
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsOneClassifier
from sklearn.svm import LinearSVC
wine_dataset = load_wine
```

```python
X, y = wine_dataset(return_X_y=True)
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=10, shuffle= True)
```

```python
X_train.shape
```

```python
y_train
```

```python
clf = OneVsOneClassifier(LinearSVC()).fit(X_train, y_train)
```

```python
clf.predict(X_test)
```

```python
(y_test == clf.predict(X_test)).astype(int).sum()/len(y_test)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, shuffle= True)
clf = OneVsOneClassifier(LinearSVC()).fit(X_train, y_train)
clf.predict(X_test)
(y_test == clf.predict(X_test)).astype(int).sum()/len(y_test)
```

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
diabets_dataset = load_diabetes
X, y = diabets_dataset(return_X_y=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, shuffle=True, random_state=30)
```

```python
regressor= LinearRegression().fit(X_train, y_train)
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
y_test
```

```python
regr = regressor.predict(X_test)
```

```python
((y_test - regr)**2).sum()/len(y_test)
```

```python
X_test[1]
```

```python
y_test.shape
```

```python
plt.figure(figsize=(10,5))
plt.plot(regr, 'ys', y_test, 'r*')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle=True, random_state=80)
regressor= LinearRegression().fit(X_train, y_train)
regr1 = regressor.predict(X_test)
((y_test - regr1)**2).sum()/len(y_test)
```

```python
plt.figure(figsize=(10,5))
plt.plot(regr1, 'ys', y_test, 'r*')
```
