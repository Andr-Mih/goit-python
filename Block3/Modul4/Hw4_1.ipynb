```python
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
```

```python
time = np.linspace(0, 11, 12)
```

```python
plt.scatter(time, speed)
```

```python
f =interp1d(time, speed, kind = 'cubic')
more_time = np.linspace(0, 11, 10000)
```

```python
plt.plot(more_time, f(more_time))
```

```python

quad(lambda x: f(x), 0, 11)
```

```python
f1 = interp1d(time, speed, kind='quadratic')
f1(9.75), f(9.75)
```

```python
plt.plot(more_time, f1(more_time))
```

```python
quad(lambda x: f1(x), 0, 11)
```

```python
(f(more_time)-f1(more_time)).max()
```
