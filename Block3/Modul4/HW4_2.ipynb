```python
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
alpha = 0.5
N = 1000000
S0 = 990000
t0, tf = 0, 25
beta = 0.3
I0=7000
R0 = 3000
```

```python
def dsdt(t, s):
    return -alpha * s
```

```python
soln = solve_ivp(dsdt, (t0, tf), [S0], dense_output=True)
```

```python
st, sy = soln.t, soln.y[0]
z = soln.sol
plt.plot(st, sy, 'x', color='b')
```

```python
def didt(t, i):
    return z(t)-beta*i
```

```python
soln1 = solve_ivp(didt, (t0, tf), [I0], dense_output=True)
it, iy = soln1.t, soln1.y[0]
k = soln1.sol
plt.plot(it, iy, color='g', label='solve_ivp')
plt.show()
```

```python
R = N - k(t) - z(t)
```

```python
plt.plot(t, R[0], color='r', label='solve_ivp')
```

```python
fig, axs = plt.subplots(3, 1)
fig.figsize = (40,40)
axs[0].plot(st, sy, 'r', label = 'S')
axs[1].plot(it, iy, 'g', label = 'I')
axs[2].plot(t, R[0], label = 'R')
```
